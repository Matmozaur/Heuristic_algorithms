# Problem plecakowy

Optymalne rozwiązanie 1458

```python
def P(X,Wartości):
    return sum([a*b for a,b in zip(X,Wartości)])

def W(X,Wagi):
    return sum([a*b for a,b in zip(X,Wagi)])

def stan_plecaka(X,Przedmioty):
    plecak=[]
    for i in range(len(X)):
        if X[i]:
            plecak.append(Przedmioty[i])
    return plecak
```

```python
Przedmioty=['Woda','Chleb','Pochodnia','Szynka','Łuk','Lina','Talerz','Kubek','Termos','Nóż','Latarka','Kij','Buty','Miecz','Włócznia']
```

```python
Wagi=[70,73,77,80,82,87,90,94,98,106,110,113,115,118,120]
```

```python
Maxymalna_waga=750
```

Symulowane wyżarzanie

```python
best=[]
```

```python
def symulowane_wyżarzanie(Wartości,Wagi,M,cmax,T0,a):
    c=0
    T=T0
    X=[0]*len(Wagi)
    Aktualna_waga=0
    Xbest=X.copy()
    while(c<=cmax):
        j=randrange(len(Wagi))
        Y=X.copy()
        Y[j]=1-X[j]
        if(Y[j]==1 and Aktualna_waga+Wagi[j]>M):
            Y=None
        if(Y != None):
            if(Y[j]==1):
                X=Y.copy()
                Aktualna_waga=Aktualna_waga+Wagi[j]
                if(P(X,Wartości)>P(Xbest,Wartości)):
                    Xbest=X.copy()
            else:
                r=random.uniform(0, 1)
                if(r<math.exp(-(Wartości[j]/T))):
                    X=Y.copy()
                    Aktualna_waga=Aktualna_waga-Wagi[j]
        c=c+1
        T=T*a
        
        print(str(c)+'\n'+str(stan_plecaka(X,Przedmioty))+'\n'+str(P(X,Wartości)))
        print('_______________________________________')
        best.append(P(X,Wartości))
        
    print('______________________________________________________________________')
    print(stan_plecaka(Xbest,Przedmioty))
    return P(Xbest,Wartości)
        
```

```python
symulowane_wyżarzanie(Wartości,Wagi,Maxymalna_waga,cmax=100,T0=10000,a=0.9999)
```

```python
best=[]
```

```python
def find_max_i(N,X,Wagi,Wartości):
    a=-99999999
    b=-1
    for i in N:
        if math.pow(-1,X[i])*Wartości[i]/Wagi[i]>a:
            a=math.pow(-1,X[i])*Wartości[i]/Wagi[i]
            b=i
    return b
```

```python
def przeszukiwanie_tabu(Wartości,Wagi,Maxymalna_waga,cmax=100,L=1):
    c=1
    X=[0]*len(Wagi)
    Aktualna_waga=0
    Xbest=X.copy()
    tabu=[]
    while(c<=cmax):
        N=[i for i in range(len(Wagi))]
        start=max(0,c-L)
        for j in range(start,c-1):
            N.remove(tabu[j])
        for i in N.copy():
            if(X[i]==0 and (Aktualna_waga+Wagi[i]>Maxymalna_waga)):
                N.remove(i)
        print(N)
        if not N:
            break;
        i=find_max_i(N,X,Wagi,Wartości)
        tabu.append(i)
        X[i]=1-X[i]
        if(X[i]==1):
            Aktualna_waga=Aktualna_waga+Wagi[i]
        else:
            Aktualna_waga=Aktualna_waga-Wagi[i]
        if(P(X,Wartości)>P(Xbest,Wartości)):
            Xbest=X.copy()
        c=c+1
        print(str(c)+'\n'+str(stan_plecaka(X,Przedmioty))+'\n'+str(P(X,Wartości)))
        print('_______________________________________')
        best.append(P(X,Wartości))
        
    print('______________________________________________________________________')
    print(stan_plecaka(Xbest,Przedmioty))
    return P(Xbest,Wartości)
```

# Problem Komiwojażera

```python
def cost(x,G):
    c=0
    for i in range(len(x)-1):
        c=c+G[x[i]][x[i+1]]
    c=c+G[x[len(x)-1]][x[0]]
    return c

def gain(x,i,j,G):
    if(j<len(x)-1):
        c = G[x[i]][x[i+1]] + G[x[j]][x[j+1]] - G[x[i+1]][x[j+1]] - G[x[i]][x[j]]
    if(j==len(x)-1):
        c = G[x[i]][x[i+1]] + G[x[j]][x[0]] - G[x[i+1]][x[0]] - G[x[i]][x[j]]
    return c

def sortuj(P,G):
    P.sort(key=lambda a: cost(a,G))
    return P
```

```python
G=[[0 for x in range(50)] for y in range(50)]
for i in range(50):
    for j in range(50):
        G[i][j]=randrange(100)
        G[j][i]=G[i][j]
        if(i==j):
            G[i][j]=0
```

```python
def hn(x,G):
    done=False
    while done==False:
        done=True
        g0=0
        for i in range(len(x)-2):
            for j in range (i+2,len(x)):
                g=gain(x,i,j,G)
                if(g>g0):
                    g0=g
                    i0=i
                    j0=j
        if(g0>0):
            xn=x.copy()
            for a in range(len(x)):
                if(a<=i0):
                    xn[a]=x[a]
                elif(a>i0):
                    if(a<=j0):
                        xn[a]=x[(j0+i0+1-a)]
                    elif(a>j0):
                        xn[a]=x[a]
            x=xn
            done=False
    return x
                        
```

```python
def wybierz_populacje(pp,G):
    P=[0]*pp
    for i in range(pp):
        p=random.sample(range(len(G)),len(G))
        P[i]=hn(p,G)
    return P
```

```python
def rekombinacja(x,y,G,h=None):
    if(h==None):
        h=randrange(10,len(x)/2)
    j=randrange(len(x))
    T=set()
    d=[0]*len(x)
    for i in range(h):
        d[i]=y[(i+j)%len(y)]
        T.add(d[i])
    i=h
    for j in range(len(y)):
        if(x[j] not in T):
            d[i]=x[j]
            i=i+1
    d=hn(d,G)
    
    j=randrange(len(x))
    T=set()
    c=[0]*len(x)
    for i in range(h):
        c[i]=x[(i+j)%len(y)]
        T.add(c[i])
    i=h
    for j in range(len(y)):
        if(y[j] not in T):
            c[i]=y[j]
            i=i+1
    c=hn(c,G)
    return c,d
```

```python
def genetyczny_kom(pp,cmax,G,h=None):
    c=1
    P=wybierz_populacje(pp,G)
    P=sortuj(P,G)
    xbest=P[0]
    cbest=cost(P[0],G)
    while(c<=cmax):
        for i in range(int(pp/2-1)):
            P=P+list(rekombinacja(P[2*i],P[2*i +1],G,h))
        P=sortuj(P,G)
        ccurr=cost(P[0],G)
        if(ccurr<cbest):
            xbest=P[0]
            cbest=ccurr
        c=c+1
    return (xbest,cbest)
```

```python
genetyczny_kom(4,10,G,h=2)
```
